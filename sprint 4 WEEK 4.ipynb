```python
import pymongo
import pandas as pd
import numpy as np
```

```python
# CREATE A MONGO DATABASE AND NAME IT DATA TRACKER


myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["DataTracker"]
mycol = mydb["MAKE"]

# ADD PRINT PRODUCT FILE
df = pd.read_csv('CarData.csv')

MODEL = df.to_dict(orient= 'records')
x = mycol.insert_many(MODEL)
for x in mycol.find():
        print(x)
```

```python
# Delete

myclient = pymongo.MongoClient("mongodb://localhost:27017/")   

mydb = myclient["DataTracker"]                       

mycol = mydb["MAKE"]

myquery = {"TOYOTA": "SUV"}
myquery = {"KIA": "SUV"}
mycol.delete_many(myquery)
```

```python
#FILTER

import pymongo

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["DataTracker"]
mycol = mydb["MAKE"]
MAKE= list(mycol.find())

for SUV in MAKE:
    print('{0}{1}{2}{3}{4}'.format(SUV['MAKE'],SUV['MAKE'],SUV['MAKE'],SUV['MAKE'],SUV['MAKE']))
```

```python
#Update


import pymongo

myclient = pymongo.MongoClient("mongodb://localhost:27017/")

mydb = myclient["DataTracker"]

mycol = mydb["MAKE"]

myquery = { "MAKE": "TOYOTA" }
newvalues ={"$set": {"BUCCATI": "SUV"}}

mycol.update_one(myquery, newvalues)

for x in mycol.find():
    print(x)
```
